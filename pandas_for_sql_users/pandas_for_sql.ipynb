```python
import pandas as pd 
import numpy as np 

iris = pd.read_csv('../data/iris.csv')
iris
```

```python
#select in pandas
#select col1, col 2 from iris, can be just called by using double square brackets and name of columns
iris[['SepalLengthCm','PetalLengthCm','Species']]
```

```python
#where in pandas
# we can pass a list or use boolean operations
iris[iris['Species']=='Iris-virginica']
```

```python
#Group by 
# in pandas we can use groupby() funcion on the dataframe
iris.groupby('Species').size()
```

```python
#group by continues
# we can also apply different functions to other columsn too
iris.groupby('Species')['SepalLengthCm'].sum()
```

```python
iris.groupby('Species')['SepalLengthCm'].mean()
```

```python
# sorting in pandas
# we use sort_values on the columns
iris.sort_values('SepalLengthCm')
```
